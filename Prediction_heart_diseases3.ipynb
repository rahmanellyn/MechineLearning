```python
#import Library
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# masukkan  dataset
X = np.array([[25,3000,12],[34,4500,24],[29,5000,26],
             [45,6000,6],[32,5500,18],[50,7000,48],
             [40,4000,8],[28,3200,14],[35,4800,30],[22,2500,10]])
y = np.array([0,1,0,1,0,0,1,0,1,0])
```

```python
#import fungsi svm
model = SVC(kernel='linear')
 # Latih model menggunakan dataset yang sudah disiapkan
model.fit(X,y)
```

```python
#melihat supoort vector
print (f'support vector: \n{model.support_vectors_}')

print (f'support vector pada masing-masing kelas: \n{model.n_support_}')

```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y)
w = model.coef_[0]
b = model.intercept_[0]

x_points = np.linspace(1,50)
y_points = -(w[0]*x_points + b)/w[1]

plt.plot(x_points,y_points)
plt.show()
```

```python
#pengujian data baru
#masukan data uji kedalam dua variabel
X_uji= np.array([35, 4700, 23])
```

```python
prediksi = model.predict([X_uji])
print(f'hasil prediksi: {prediksi}')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('heart_diseases.csv')
data.head()
```

```python
#tentukan fitur dan target
X = data.drop('num', axis=1) #fitur
y = data['num'] #target
```

```python
#bagi data latih dan data ujii

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
model2 = SVC()
model2 = model2.fit(X_train, y_train)
model2
```

```python
#lakukan prediksi
model2_pred = model2.predict(X_test)
model2_pred
```

```python
#akurasi model
from sklearn.metrics import accuracy_score
accuracy_score(y_test, model2_pred)
```
